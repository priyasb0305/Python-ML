**Data Preprocessing**

- Clean column names

```python
df.columns=df.columns.str.strip()
```

- Convert Entry column to datetime

```python
df['Entry_Date']=pd.to_datetime(df['Entry'],format='%b/%Y')
df['Entry_Year']=df['Entry_Date'].dt.year
df.head()
```

```python
df.dtypes
```

- Create Valuation growth metric

```python
df['Valuation_Growth']=df['Valuation ($B)']/df['Entry Valuation^^ ($B)']
df.head()
```

- Extract primary city(before '/')

```python
df['Primary_City']=df['Location'].str.split('/').str[0]
df.head()
```

- Count numbers of Investors

```python
df['Investor_Count']=df['Select Investors'].str.count(',')+1
df.head()
```

# Univariate Analysis

- Valuation Distribution Analysis

```python
plt.hist(df['Valuation ($B)'],bins=20,alpha=0.7,color='pink',edgecolor='red')
plt.title('Distribution of Current Valuations')
plt.xlabel('Valuation ($B)')
plt.ylabel('Frequency')
plt.show()
```

# Sector Analysis

 - Sector distribution

```python
plt.figure(figsize=(12,8))

sector_counts=df['Sector'].value_counts()
colors=plt.cm.Set2(np.linspace(0,1,len(sector_counts)))

plt.pie(sector_counts.values,labels=sector_counts.index,colors=colors,autopct='%1.1f%%',startangle=0)
plt.title('Distribution of Unicorns by Sector', fontsize=16,fontweight='bold')
plt.axis('equal')
plt.show()
```

```python
def display_top_k_sectors(df,k):
  sector_counts=df['Sector'].value_counts()

  #Select top K sectors and group the rest as 'others'
  if k<len(sector_counts):
    top_sectors=sector_counts.head(k)
    other_sectors_count=sector_counts.iloc[k:].sum()
    top_sectors['Others']=other_sectors_count
  else:
    top_sectors=sector_counts

  plt.figure(figsize=(12,8))
  colors=plt.cm.Set2(np.linspace(0,1,len(top_sectors)))

  plt.pie(top_sectors.values,labels=top_sectors.index,colors=colors,autopct='%1.1f%%',startangle=0)
  plt.title(f'Distribution of Top {k} Unicorns by Sector', fontsize=16,fontweight='bold')
  plt.axis('equal')
  plt.show()

#call the function with K=5 to display top 5 sectors
display_top_k_sectors(df,5)
```

- Top selector bar chart

```python
plt.figure(figsize=(12,6))
top_sectors=sector_counts.head(10)
bars=plt.bar(range(len(top_sectors)),top_sectors.values,color='pink')
plt.title('Top 10 Sectors by Number of Unicorns',fontsize=14,fontweight='bold')
plt.xlabel('Sector')
plt.ylabel('Number of Unicorns')
plt.xticks(range(len(top_sectors)),top_sectors.index,rotation=45,ha='right')

#Add value labels on bars
for bar,value in zip(bars,top_sectors.values):
  plt.text(bar.get_x()+bar.get_width()/2,bar.get_height() - 0.8,str(value),ha='center',va='bottom')

plt.tight_layout()
plt.show()
```

Key Insights:
 - EdTech leads with highest number of unicorns
 - E-Commerce and SaaS are major contributors
 - Edtech emergence as significant sector
 - Diversification across multiple sectors

# Geographic Distribution

- City-Wise distribution

```python
plt.figure(figsize=(14,6))

city_counts=df['Primary_City'].value_counts().head(10)
bars=plt.bar(city_counts.index,city_counts.values,color='aqua')
plt.title('Top 10 Cities by Number of Unicorns',fontsize=14,fontweight='bold')
plt.xlabel('Cities')
plt.ylabel('Number of Unicorns')
plt.xticks(rotation=45)

#Add value labels
for bar,values in zip(bars,city_counts.values):
  plt.text(bar.get_x()+bar.get_width()/2,bar.get_height()+0.1,str(values),ha='center',va='bottom')

plt.tight_layout()
plt.show()
```

- Create a summary Table

```python
city_summary=df.groupby('Primary_City').agg({'Valuation ($B)':['count','sum','mean'],
                                             'Entry_Year':['min','max']}).round(2)
city_summary.columns=['Count','Total_Valuation','Avg_Valuation','First Entry','Last Entry']
city_summary.head(10)
```

Key Insights:
- Banglore Dominates with highest number of unicorns
- Mumbai and Gurgaon are other major hubs
- Geographic concentration in tech cities
- Banglore leads in both count and total valuation

**Temporal Analysis**

- Entry Year Trends

```python
yearly_entries=df['Entry_Year'].value_counts().sort_index()
bars=plt.bar(yearly_entries.index,yearly_entries.values,color='teal')
plt.title('Number of Unicorns Entered by Year',fontsize=14,fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Number of Unicorns')
plt.xticks(yearly_entries.index,rotation=45)

#Add value labels
for bar,value in zip(bars,yearly_entries.values):
  plt.text(bar.get_x()+bar.get_width()/2,bar.get_height()+0.1,str(value),ha='center',va='bottom')

plt.tight_layout()
plt.show()
```

- Cumulative unicorns

```python
cumulative=yearly_entries.cumsum()
plt.plot(cumulative.index,cumulative.values,marker='o',linestyle='--',color='blue')
plt.title('Cumulative Unicorns over Time',fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Cumulative Unicorns')
plt.grid(True,alpha=0.3)
plt.show()
```

- Average valuation by year

```python
avg_val_by_year=df.groupby('Entry_Year')['Valuation ($B)'].mean()
plt.bar(avg_val_by_year.index,avg_val_by_year.values,color='olive')
plt.title('Average Valuation by Year',fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Average Valuation ($B)')
plt.show()
```

- Sector evolution over time

```python
def display_k_sector_evolution(df,k):
  #Group by year and sector,count occurrences
  sector_year=df.groupby(['Entry_Year','Sector']).size().unstack(fill_value=0)

  #Select top K sectors based on overall count
  top_k_sectors_names=df['Sector'].value_counts().head(k).index.tolist()

  sector_year=sector_year[top_k_sectors_names]

  plt.figure(figsize=(15,8))
  sector_year.plot(kind='bar',alpha=0.7,ax=plt.gca())

  plt.title(f'Evolution of Top {k} Sectors over Time',fontsize=14,fontweight='bold')
  plt.xlabel('Year')
  plt.ylabel('Number of Unicorns')
  plt.legend(title='Sector',bbox_to_anchor=(1.05,1),loc='upper left')
  plt.yticks(range(0,sector_year.values.max()+5))
  plt.grid(True,alpha=0.3)
  plt.tight_layout()
  plt.show()

#call the function with K=5 to display top 5 sectors
display_k_sector_evolution(df,5)
```
