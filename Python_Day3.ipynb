# **Numpy**

Create Arrays

```python
import numpy as np

a=np.array([8,7,6,5,4])
b=np.arange(1,5)
c=np.zeros((2,3))
d=np.ones((3,2))

print("Array a:",a)
print("Array b:",b)
print("2x3 Zeros:",c)
print("3x2 Ones:",d)
```

- Indexing and slicing

```python
print("First element of a:",a[0])
print("Last two elements of a:",a[-2:])
```

- Basic arithmetic operations

```python
print("a+2:",a+2)
print("a*3:",a*3)
```

- Basic statistics


```python
print("Square of a:",np.square(a))
print("Mean of a:",np.mean(a))
print("Standard diviation:",np.std(a))
```

- Reshaping arrays

```python
matrix=np.arange(1,10).reshape((3,3))
print("Reshaped 3x3 matrix:")
matrix
```

- Transpose

```python
print("Transpose:")
matrix.T
```

- Boolean indexing

```python
print("Elements<=5:",matrix[matrix<=5])
```

# **Pandas**

Create a DataFrame manually

```python
import pandas as pd

data={
    'Name':['Alice','Bob','Charlie','david','Eva'],
    'Age':[25,30,35,40,22],
    'Department':['Sales','Engineering','HR','Engineering','Sales'],
    'Salary':[50000,80000,40000,90000,45000]
}
df=pd.DataFrame(data)
df
```

- Basic operations

```python
print("First 3 rows:")
df.head(3)
```

```python
print("Column name:")
df.columns
```

```python
print("Data Types:")
df.dtypes
```

```python
print("Describe numerical columns:")
df.describe()
```

- Filtering

```python
print("People with salary >=50000:")
df[df['Salary']>=50000]
```

- Sorting


```python
print("Sorted by Age:")
df.sort_values(by='Age')
```

- Groupping and aggregation

```python
grouped=df.groupby('Department')['Salary'].mean()
print("Average salary by Department:")
grouped
```

- Adding new columns

```python
df['Salary in Lakhs']=df['Salary']/100000
df
```

# **Matplotlib**

**Line Plot**

```python
import matplotlib.pyplot as plt

years=[2019,2020,2021,2022]
revenue=[100,220,130,160]

plt.plot(years,revenue,marker='^',linestyle='--',color='red')
plt.title("Company Revenue over years")
plt.xlabel("Years")
plt.ylabel("Revenue(in crores)")
plt.xticks(years)
plt.grid(True)
plt.show()
```

**Bar chart**

```python
departments=df['Department'].value_counts()
departments.plot(kind='bar',color='beige')
plt.title("Department Distribution")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

**Histogram**

```python
ages=[33,18,38,45,10,50]
plt.hist(ages,bins=5,color='aqua',edgecolor='blue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

**Pie chart**

```python
df.groupby('Department').size().plot(kind='pie',autopct='%1.1f%%')
plt.title("Department Proportion")
plt.ylabel("")
plt.show()
```

# **Seaborn**

- Load sample dataset

```python
import seaborn as sns

tips=sns.load_dataset("tips")
tips.head()
```

```python
print("\nInfo about the tips dataset:")
tips.info()

print("\nDescription of the tips dataset:")
tips.describe(include='all')
```

```python
sns.pairplot(tips,hue="sex")
plt.show()
```

```python
corr=tips.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("correlation matrix")
plt.show()
```

```python
sns.countplot(x='day',data=tips,palette='Set3',hue='day',legend=True)
plt.title("Count of records by day")
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Boxplot of Total Bill by Day")
plt.show()
```
